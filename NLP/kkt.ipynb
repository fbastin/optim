# Conditions de Karush-Kuhn-Tucker

## Exercice 1

*Adapté d'un exercice de B. Des Graupes, M1 Méthodes Numériques, Université Paris Nanterre.*

Nous considérons une économie à deux produits: une bien de consommation et du travail. L'indice 1 désigne le bien de consommation, avec un prix $p_1  > 0$, tandis que l'indice 2 désigne le travail, avec un salaire $𝑝_2> 0$.

Les préférences d'un agent économique, avec comme seule ressource la main-d'œuvre, peuvent être représentées avec une fonction d'utilité
$$
U(x_1, x_2) = 2 \ln x_1 + \ln (3 - x_2).
$$

On suppose qu'il existe un niveau de consommation minimum $x_1 \geq 1$ alors que le travail ne peut pas dépasser 3. Nous voulons déterminer la fonction de demande du bien de consommation et la fonction d'offre d'emploi. Formulez le programme de maximisation, et résolvez-le à l'aide des conditions de Karush-Kuhn-Tucker. La solution est-elle unique? Pourquoi?

Nous devons maximiser l'utilité $U(x_1, x_2) = 2 \ln x_1 + \ln (3 - x_2)$ sous les contraintes $x_1 \geq 1$ et $x_2 \leq 3$. De plus, il n'est pas possible de dépenser plus que ce qui est gagné: $p_1x_1 \leq p_2x_2$. En mettant le tout ensemble, nous obtenons le programme

\begin{align*}
\min_x\ & -2 \ln x_1 - \ln (3 - x_2) \\
\mbox{s.t. } & x_1 \geq 1 \\ & x_2 \leq 3 \\
& p_1x_1 - p_2x_2 \leq 0 \\
& x_1, x_2 \geq 0.
\end{align*}

Nous pouvons reformuler le problème comme
\begin{align*}
\min_x\ & -2 \ln x_1 - \ln (3 - x_2) \\
\mbox{s.t. } & -x_1+1 \leq 0 \\ & x_2 - 3 \leq 0 \\
& p_1x_1 - p_2x_2 \leq 0 \\
& -x_1 \leq 0, -x_2 \leq 0.
\end{align*}
Dès lors, le lagrangien est
$$
L(x,\lambda) = -2 \ln x_1 - \ln (3 - x_2) + \lambda_1 (1-x_1) + \lambda_2 (x_2-3) + \lambda_3 (p_1x_1 - p_2x_2) - \lambda_4 x_1 - \lambda_5 x_2.
$$

Les conditions KKT peuvent être écrites comme
\begin{align*}
-\frac{2}{x_1}-\lambda_1+p_1\lambda_3-\lambda_4 &= 0 \\
\frac{1}{3-x_2}+\lambda_2-p_2\lambda_3-\lambda_5 &= 0 \\
-x_1+1 &\leq 0 \\
x_2 - 3 & \leq 0 \\
p_1x_1 - p_2x_2 & \leq 0 \\
-x_1 &\leq 0 \\
-x_2 & \leq 0 \\
\lambda_1(-x_1+1) & = 0 \\
\lambda_2(x_2 - 3) & = 0 \\
\lambda_3(p_1x_1 - p_2x_2) & = 0 \\
\lambda_4 x_1 & = 0 \\
\lambda_5 x_2 & = 0 \\
\lambda_i & \geq 0,\ i = 1,\ldots,5
\end{align*}

Nous ne pouvons avoir $x_1 = 0$, puisque cela violerait la contrainte $-x_1 + 1 \leq 0$, et par conséquent $\lambda_4 = 0$. De plus, la contrainte $x_1 \geq 0$ est impliquée par la contrainte $x_1 \geq 1$. Nous pouvons dès lors simplifier le système KKT comme suit:
\begin{align*}
-\frac{2}{x_1}-\lambda_1+p_1\lambda_3 &= 0 \\
\frac{1}{3-x_2}+\lambda_2-p_2\lambda_3-\lambda_5 &= 0 \\
-x_1+1 &\leq 0 \\
x_2 - 3 & \leq 0 \\
p_1x_1 - p_2x_2 & \leq 0 \\
-x_2 & \leq 0 \\
\lambda_1(-x_1+1) & = 0 \\
\lambda_2(x_2 - 3) & = 0 \\
\lambda_3(p_1x_1 - p_2x_2) & = 0 \\
\lambda_5 x_2 & = 0 \\
\lambda_i & \geq 0,\ i = 1,\ldots,5, \lambda_4 = 0.
\end{align*}

Nous pouvons aussi exclure le cas $x_2 = 0$, comme autrement, la contrainte $p_1 x_1 - p_2 x_2 \leq 0$ contredirait la contrainte $x_1 \geq 1$. Dès lors, $\lambda_5 = 0$, et le système peut être davantage simplifié comme suit:
\begin{align*}
-\frac{2}{x_1}-\lambda_1+p_1\lambda_3 &= 0 \\
\frac{1}{3-x_2}+\lambda_2-p_2\lambda_3 &= 0 \\
-x_1+1 &\leq 0 \\
x_2 - 3 & \leq 0 \\
p_1x_1 - p_2x_2 & \leq 0 \\
-x_2 & \leq 0 \\
\lambda_1(-x_1+1) & = 0 \\
\lambda_2(x_2 - 3) & = 0 \\
\lambda_3(p_1x_1 - p_2x_2) & = 0 \\
\lambda_i & \geq 0,\ i = 1,2,3.
\end{align*}

Observons aussi que $x_2 \ne 3$, comme autrement la dérivée du lagrangien par rapport à $x_2$ ne serait pas définie. Par conséquent, $x_2 < 3$, $\lambda_2 = 0$ et
$$
\frac{1}{3-x_2} = p_2\lambda_3.
$$
Ceci entraîne $\lambda_3 \ne 0$ et $p_1x_1 = p_2x_2$.

\begin{align*}
-\frac{2}{x_1}-\lambda_1+p_1\lambda_3 &= 0 \\
\frac{1}{3-x_2}-p_2\lambda_3 &= 0 \\
-x_1+1 &\leq 0 \\
x_2 - 3 & \leq 0 \\
p_1x_1 &= p_2x_2 \\
-x_2 & \leq 0 \\
\lambda_1(-x_1+1) & = 0 \\
\lambda_3(p_1x_1 - p_2x_2) & = 0 \\
\lambda_1 & \geq 0,\ \lambda_3 \geq 0.
\end{align*}

Supposons $x_1 = 1$. Alors
$$
x_2 = \frac{p_1}{p_2}.
$$
Si $\frac{p_1}{p_2} \geq 3$, la solution violerait la contrainte $x_2 < 3$, aussi nous devons exclure ce cas. Ainsi, $\frac{p_1}{p_2} < 3$, et les conditions
$$
x_2 \geq 0,\ p_1x_1 - p_2x_2 \leq 0
$$
sont trivialement satisfaites. De la seconde inégalité, nous tirons
$$
\frac{1}{3-p_1/p_2} = p_2\lambda_3,
$$
ou, de manière équivalente,
$$
\frac{1}{3p_2-p_1} = \lambda_3.
$$
Comme $\lambda_3 \geq 0$, cette égalité ne peut tenir que si $3p_2-p_1 > 0$, ou $p_1 < 3p_2$.
La première égalité donne
$$
-2 - \lambda_1+p_1 \frac{1}{3p_2-p_1} = 0
$$
ou
$$
\lambda_1 = \frac{-2(3p_2-p_1) + p_1}{3p_2-p_1}
$$
Comme nous devons avoir $\lambda_1 \geq 0$, ceci n'est satisfait que si
$$
-6p_2 + 3p_1 \geq 0
$$
ou
$$p_1 \geq 2p_2.$$

Supposons maintenant que $x_1 > 1$. Alors, $\lambda_1 = 0$ et
$$
\lambda_3 = \frac{2}{p_1x_1}.
$$
Ceci à son tour implique
$$
\frac{1}{3-x_2}-\frac{2p_2}{p_1x_1} = 0,
$$
ou
$$
p_1x_1 - 2p_2(3-x_2) = p_1x_1 +2p_2x_2 -6p_2 = p_1x_1 - p_2x_2 +3p_2x_2 -6p_2 = 0.
$$
Dès lors, $x_2 = 2$. Ceci conduit à
$$
x_1 = 2\frac{p_2}{p_1}, \ \lambda_3 = \frac{1}{p_2}.
$$
Toutes les conditions sont alors satisfaites avec la condition $2\frac{p_2}{p_1} > 1$, ou $p_1 < 2p_2$.

En résumé, nous avons trois cas à considérer
1. $p_1 < 2p_2$: $x_1 = 2p_2/p_1$, $x_2 = 2$;
2. $2p_2 \leq p_1 < 3p_2$: $x_1 = 1$, $x_2 = \frac{p_1}{p_2}$;
3. $p_1 \geq 3p_2$: pas de solution.

## Exercice 2

Considérons le problème
\begin{align*}
\min\ & f(x,y) = x^2+2y^2 \\
\mbox{t.q. } & x+y-b = 0
\end{align*}

Le lagrangien est
$$
L(\lambda,x,y) = x^2 + 2y^2 + \lambda(x+y-b)
$$

Le lagrangien est convexe en $(x,y)$ et atteint un minimum quand $\nabla_{x,y} L(\lambda, x, y) = 0$. En effet,
$$
\nabla_{(x,y)}^2 L(\lambda, x, y) = \begin{pmatrix} 2 & 0 \\ 0 & 4 \end{pmatrix}
$$
qui est définie positive (strictement diagonalement dominante).

\begin{align*}
& \nabla_{x,y} L(\lambda, x, y) =
\begin{pmatrix}
2x + \lambda \\
4y + \lambda
\end{pmatrix}
= 0 \\
\Leftrightarrow &
\begin{cases}
\lambda = -2x \\
\lambda = -4y
\end{cases}
\end{align*}
Dès lors, $\nabla_{x,y} L(\lambda, x, y) = 0$ implique $x = 2y$, et de la contrainte,
$$
2y + y - b = 0
$$
ou
$$
y = \frac{b}{3}
$$
et
$$
x = \frac{2b}{3}
$$

## Exercice 3: Conditions KKT avec contraintes mixtes

Considérons le problème convexe
\begin{align*}
\min_{x,y}\ & (x-3)^2 + (y-2)^2 \\
\mbox{t.q. } & x + y = 3 \\
& x \geq 0 \\
& y \geq 0.
\end{align*}

1. Écrire le lagrangien du problème.
2. Écrire les conditions KKT.
3. Résoudre le système KKT en considérant les différents cas de complémentarité.
4. Vérifier numériquement la solution avec JuMP.

### Solution

**1.** Reformulons les contraintes d'inégalité sous la forme $g_i(x) \leq 0$: $-x \leq 0$ et $-y \leq 0$. Le lagrangien est
$$
L(x, y, \lambda, \mu) = (x-3)^2 + (y-2)^2 + \mu(x+y-3) - \lambda_1 x - \lambda_2 y
$$
où $\mu \in \mathbb{R}$ est le multiplicateur associé à la contrainte d'égalité et $\lambda_1, \lambda_2 \geq 0$ sont les multiplicateurs associés aux contraintes d'inégalité.

**2.** Les conditions KKT sont
\begin{align*}
2(x-3) + \mu - \lambda_1 &= 0 & \text{(stationarité en } x\text{)} \\
2(y-2) + \mu - \lambda_2 &= 0 & \text{(stationarité en } y\text{)} \\
x + y &= 3 & \text{(faisabilité primale, égalité)} \\
-x &\leq 0,\quad -y \leq 0 & \text{(faisabilité primale, inégalités)} \\
\lambda_1 x &= 0 & \text{(complémentarité)} \\
\lambda_2 y &= 0 & \text{(complémentarité)} \\
\lambda_1, \lambda_2 &\geq 0 & \text{(faisabilité duale)}
\end{align*}

**3.** Nous procédons par analyse de cas.

**Cas 1: $\lambda_1 = 0$, $\lambda_2 = 0$.** De la stationarité,
$$
\mu = -2(x-3) = 6-2x, \quad \mu = -2(y-2) = 4-2y.
$$
Dès lors, $6-2x = 4-2y$, ce qui donne $y = x-1$. Avec la contrainte $x+y = 3$, on obtient $x = 2$, $y = 1$, $\mu = 2$. Les conditions $x \geq 0$ et $y \geq 0$ sont satisfaites.

**Cas 2: $\lambda_1 > 0$, $\lambda_2 = 0$.** Alors $x = 0$ et $y = 3$. La stationarité donne
$$
\mu = \lambda_1 + 2(0-3) = \lambda_1 - 6, \quad \mu = -2(3-2) = -2.
$$
Dès lors, $\lambda_1 = 4 > 0$, ce qui est cohérent. Cependant, la valeur objectif est $(0-3)^2 + (3-2)^2 = 10 > 6$.

**Cas 3: $\lambda_1 = 0$, $\lambda_2 > 0$.** Alors $y = 0$ et $x = 3$. La stationarité donne
$$
\mu = -2(3-3) = 0, \quad \mu = \lambda_2 + 2(0-2) = \lambda_2 - 4.
$$
Dès lors, $\lambda_2 = 4 > 0$. Valeur objectif: $(3-3)^2 + (0-2)^2 = 4 < 6$. Notons cependant que nous ne pouvons pas comparer directement les valeurs objectif pour décider du minimum, car dans le cas convexe, tout point KKT est un minimum global.

Vérifions: au Cas 3, $x = 3$, $y = 0$, $\mu = 0$, $\lambda_1 = 0$, $\lambda_2 = 4$. Toutes les conditions KKT sont satisfaites, et le problème est convexe. La valeur objectif est $4$.

Au Cas 1, $x = 2$, $y = 1$, $\mu = 2$, $\lambda_1 = \lambda_2 = 0$. Toutes les conditions KKT sont aussi satisfaites. La valeur objectif est $(2-3)^2 + (1-2)^2 = 2$.

Le minimum global est donc au Cas 1: $x^* = 2$, $y^* = 1$, avec $f^* = 2$.

Mais attendons — dans le cas convexe, les conditions KKT sont nécessaires et suffisantes pour l'optimalité globale. Comment se fait-il qu'il y ait plusieurs points KKT?

En fait, au Cas 2 et au Cas 3, nous avons trouvé des points KKT qui ne sont pas des minima globaux. Ceci est impossible si le problème est convexe. Vérifions: au Cas 3, la valeur objectif vaut $4 > 2$, donc $(3,0)$ n'est pas optimal. Ceci signifie que nous avons commis une erreur, ou bien que les conditions de qualification de contraintes ne tiennent pas.

En réalité, dans un programme convexe, tout point KKT est un minimum global, et la valeur objectif est la même en tout point KKT. Revérifions le Cas 3: $\nabla_x L = 2(3-3) + 0 - 0 = 0$ ✓, $\nabla_y L = 2(0-2) + 0 - 4 = -8 \ne 0$ ✗. Le Cas 3 ne satisfait pas la stationarité! Nous avions omis un signe. Reprenons: $\nabla_y L = 2(y-2) + \mu - \lambda_2 = 2(0-2) + 0 - \lambda_2 = -4-\lambda_2$, qui ne peut s'annuler pour $\lambda_2 \geq 0$. Le Cas 3 n'admet donc pas de solution KKT.

Le même raisonnement s'applique au Cas 2. Revérifions: $\nabla_x L = 2(0-3) + \mu - \lambda_1 = -6 + \mu - \lambda_1$. Avec $\mu = -2$ et $\lambda_1 = 4$: $-6 -2 -4 = -12 \ne 0$. Le Cas 2 est aussi invalide.

La seule solution KKT est donc bien $(x^*, y^*) = (2, 1)$, avec $\mu^* = 2$ et $\lambda^* = 0$.

**4.** Vérification numérique.

```julia
using JuMP, Ipopt

model = Model(Ipopt.Optimizer)
set_silent(model)

@variable(model, x >= 0)
@variable(model, y >= 0)
@constraint(model, x + y == 3)
@objective(model, Min, (x - 3)^2 + (y - 2)^2)

optimize!(model)

println("x* = ", value(x))
println("y* = ", value(y))
println("f* = ", objective_value(model))
```

## Exercice 4: Projection sur contraintes linéaires d'égalité via KKT

Soit $A = \begin{pmatrix} 1 & 1 & 1 \end{pmatrix}$, $b = 1$ et $y = \begin{pmatrix} 2 \\ 3 \\ 1 \end{pmatrix}$.

1. Calculer la projection de $y$ sur l'ensemble $\mathcal{X} = \{ x \in \mathbb{R}^3 \mid Ax = b \}$ en résolvant les conditions KKT du problème
$$
\min_x \frac{1}{2} \| y - x \|^2 \quad \text{t.q.} \quad Ax = b.
$$
2. Calculer le gradient projeté $\bar{d} = -(I - A^T(AA^T)^{-1}A)\nabla f(x)$ pour $f(x) = x_1^2 + x_2^2 + x_3^2$ au point $x = \begin{pmatrix} 1/3 \\ 1/3 \\ 1/3 \end{pmatrix}$.
3. Vérifier que $A\bar{d} = 0$.

### Solution

**1.** Avec $H = I$, les conditions KKT du problème de projection donnent le système
$$
\begin{pmatrix} I & A^T \\ A & 0 \end{pmatrix}
\begin{pmatrix} x \\ -\lambda \end{pmatrix}
= \begin{pmatrix} y \\ b \end{pmatrix}.
$$
De la première équation, $x = y + A^T\lambda$. En substituant dans la seconde:
$$
A(y + A^T\lambda) = b \implies AA^T\lambda = b - Ay.
$$
Or $AA^T = 1^2 + 1^2 + 1^2 = 3$ et $Ay = 2 + 3 + 1 = 6$. Dès lors,
$$
3\lambda = 1 - 6 = -5 \implies \lambda = -\frac{5}{3}.
$$
Puis
$$
x = y + A^T\lambda = \begin{pmatrix} 2 \\ 3 \\ 1 \end{pmatrix} + (-5/3) \begin{pmatrix} 1 \\ 1 \\ 1 \end{pmatrix} = \begin{pmatrix} 1/3 \\ 4/3 \\ -2/3 \end{pmatrix}.
$$
Vérifions: $Ax = 1/3 + 4/3 - 2/3 = 1$ ✓.

**2.** La matrice de projection sur le noyau de $A$ est
$$
P = I - A^T(AA^T)^{-1}A = I - \frac{1}{3}\begin{pmatrix} 1 \\ 1 \\ 1 \end{pmatrix}\begin{pmatrix} 1 & 1 & 1 \end{pmatrix} = I - \frac{1}{3}\mathbf{1}\mathbf{1}^T.
$$
Au point $x = (1/3, 1/3, 1/3)^T$, $\nabla f(x) = 2x = (2/3, 2/3, 2/3)^T$. Dès lors,
$$
\bar{d} = -P\nabla f(x) = -\left(I - \frac{1}{3}\mathbf{1}\mathbf{1}^T\right)\begin{pmatrix} 2/3 \\ 2/3 \\ 2/3 \end{pmatrix} = -\begin{pmatrix} 2/3 \\ 2/3 \\ 2/3 \end{pmatrix} + \frac{1}{3}\begin{pmatrix} 1 \\ 1 \\ 1 \end{pmatrix}\frac{2}{1} = \begin{pmatrix} 0 \\ 0 \\ 0 \end{pmatrix}.
$$
Le gradient projeté est nul: $x = (1/3, 1/3, 1/3)^T$ est un point stationnaire du problème $\min \|x\|^2$ sous $Ax = b$, ce qui est le minimum global puisque le problème est convexe.

**3.** $A\bar{d} = A \cdot 0 = 0$ ✓. Plus généralement, pour toute direction $d$, $AP d = A(I - A^T(AA^T)^{-1}A)d = Ad - Ad = 0$.

```julia
using LinearAlgebra

A = [1.0 1.0 1.0]
b = 1.0
y = [2.0, 3.0, 1.0]

# Projection via KKT
λ = (b - (A * y)[1]) / (A * A')[1]
x_proj = y + A' * λ
println("Projection: x = ", x_proj)
println("Vérification Ax = ", (A * x_proj)[1])

# Gradient projeté
P = I - A' * inv(A * A') * A
x0 = [1/3, 1/3, 1/3]
grad_f = 2 * x0
d_bar = -P * grad_f
println("Gradient projeté: d̄ = ", d_bar)
println("Vérification Ad̄ = ", (A * d_bar)[1])
```

## Exercice 5: Ensemble actif, LICQ et suffisance des conditions KKT

Considérons le problème
\begin{align*}
\min_{x,y}\ & (x-4)^2 + (y-4)^2 \\
\mbox{t.q. } & x + y \leq 4 \\
& x \leq 3 \\
& x, y \geq 0.
\end{align*}

1. Identifier l'ensemble actif $\mathcal{A}(x^*)$ au point $x^* = (2, 2)$. La LICQ est-elle satisfaite?
2. Écrire et résoudre les conditions KKT.
3. Le problème est-il convexe? Que pouvez-vous conclure quant à la nature de la solution?
4. Vérifier numériquement.

### Solution

**1.** Reformulons les contraintes sous la forme $g_i(x,y) \leq 0$:
$$
g_1(x,y) = x + y - 4,\quad g_2(x,y) = x - 3,\quad g_3(x,y) = -x,\quad g_4(x,y) = -y.
$$
Au point $(2,2)$: $g_1 = 0$, $g_2 = -1 < 0$, $g_3 = -2 < 0$, $g_4 = -2 < 0$. L'ensemble actif est donc $\mathcal{A}(x^*) = \{1\}$.

Le gradient de la seule contrainte active est $\nabla g_1 = (1, 1)^T$, qui est non nul. L'ensemble $\{\nabla g_1(x^*)\}$ est linéairement indépendant: la LICQ est satisfaite.

**2.** Le lagrangien est
$$
L(x, y, \lambda) = (x-4)^2 + (y-4)^2 + \lambda_1(x+y-4) + \lambda_2(x-3) - \lambda_3 x - \lambda_4 y.
$$
Les conditions KKT sont
\begin{align*}
2(x-4) + \lambda_1 + \lambda_2 - \lambda_3 &= 0 \\
2(y-4) + \lambda_1 - \lambda_4 &= 0 \\
\lambda_1(x+y-4) &= 0,\quad \lambda_2(x-3) = 0 \\
\lambda_3 x &= 0,\quad \lambda_4 y = 0 \\
\lambda_i &\geq 0,\ i = 1,\ldots,4.
\end{align*}

Essayons le cas où seule $g_1$ est active ($\lambda_2 = \lambda_3 = \lambda_4 = 0$, $\lambda_1 > 0$). Alors $x + y = 4$, et
$$
2(x-4) + \lambda_1 = 0, \quad 2(y-4) + \lambda_1 = 0.
$$
De ces deux équations, $x = y$, et avec $x + y = 4$, on obtient $x = y = 2$. Alors $\lambda_1 = -2(2-4) = 4 > 0$ ✓.

On vérifie la faisabilité: $x + y = 4$ ✓, $x = 2 \leq 3$ ✓, $x, y \geq 0$ ✓.

**3.** La fonction objectif est une somme de carrés, donc convexe. Les contraintes sont linéaires, donc l'ensemble réalisable est convexe. Le problème est convexe et la condition de Slater est satisfaite (par exemple, $(1,1)$ vérifie toutes les inégalités strictement). Par conséquent, les conditions KKT sont nécessaires et suffisantes pour l'optimalité globale. La solution $(2,2)$ est le minimum global, avec $f^* = (2-4)^2 + (2-4)^2 = 8$.

**4.** Vérification numérique.

```julia
using JuMP, Ipopt

model = Model(Ipopt.Optimizer)
set_silent(model)

@variable(model, x >= 0)
@variable(model, y >= 0)
@constraint(model, x + y <= 4)
@constraint(model, x <= 3)
@objective(model, Min, (x - 4)^2 + (y - 4)^2)

optimize!(model)

println("x* = ", value(x))
println("y* = ", value(y))
println("f* = ", objective_value(model))
```

```julia
# Vérification de la LICQ
using LinearAlgebra

# Gradients des contraintes actives en (2,2)
# Seule g1 est active
∇g1 = [1.0, 1.0]
println("Gradient de la contrainte active: ", ∇g1)
println("Rang de la matrice des gradients actifs: ", rank(reshape(∇g1, 1, 2)))
println("LICQ satisfaite: ", rank(reshape(∇g1, 1, 2)) == 1)
```

## Contre-exemples aux conditions LICQ et de Slater

Les sections qui suivent illustrent des situations où les conditions de qualification de contraintes (LICQ, Slater) ne sont pas satisfaites, et les conséquences que cela entraîne sur les conditions KKT et la dualité.

### Exemple 1

#### LICQ

Considérons le problème d'optimisation
\begin{align*}
\min_{x,y} \ & y \\
\mbox{s.t. } & (x-1)^2 + y^2 = 1 \\
& (x+1)^2 + y^2 = 1
\end{align*}

Si nous soustrayons la seconde contrainte à la première, nous obtenons
$$
(x-1)^2 - (x+1)^2 = 0
$$
ou, de manière équivalente,
$$
x^2-2x+1 - x^2-2x-1 = -4x = 0
$$
Dès lors, $x = 0$, et par conséquent, $y = 0$.
L'ensemble réalisable est dès lors le singleton $\{ (0,0) \}$

La matrice jacobienne des contraintes est
$$
J(x,y) =
\begin{pmatrix}
2(x-1) & 2y \\
2(x+1) & 2y \\
\end{pmatrix}
$$
Dès lors,
$$
J(0,0) =
\begin{pmatrix}
-2 & 0 \\
2 & 0 \\
\end{pmatrix}
$$
Les gradients des contraintes actives sont linéairement dépendants. En d'autres termes, la LICQ n'est pas satisfaite.

```julia
using LinearAlgebra

J = [ -2 0 ; 2 0]
rank(J)
```

Le lagrangien est
$$
L(x, y, \lambda) = y + \lambda_1 ((x-1)^2 + y^2 - 1) + \lambda_2 ((x + 1)^2 + y^2 - 1)
$$
et les conditions KKT sont
\begin{align*}
2\lambda_1 (x-1) + 2\lambda_2 (x+1) &= 0 \\
1 + 2\lambda_1 y + 2\lambda_2 y &= 0 \\
(x-1)^2 + y^2 &= 1 \\
(x+1)^2 + y^2 &= 1
\end{align*}
Puisque seulement (0,0) est réalisable, les deux premières égalités deviennent
\begin{align*}
-2\lambda_1 + 2\lambda_2 &= 0 \\
1 &= 0
\end{align*}
ce qui est évidemment faux.

Nous pouvons établir les conditions de dualité en considérant le problème
\begin{align*}
q(\lambda) &= \inf_{(x,y)} \{y + \lambda_1 ((x-1)^2 + y^2 - 1) + \lambda_2 ((x + 1)^2 + y^2 - 1) \} \\
 &= \inf_{(x,y)} \{y + \lambda_1 ( x^2 - 2x + y^2 ) + \lambda_2 (  x^2 + 2x + y^2 ) \}
\end{align*}
avec $\lambda = (\lambda_1, \lambda_2)$.

Si $\lambda = 0$, $q(0) = -\infty$.

Si $\lambda_1 = -\lambda_2$. Alors
\begin{align*}
q(\lambda) &= \inf_{(x,y)} \{y + \lambda_1 ((x-1)^2 + y^2 - 1) - \lambda_1 ((x + 1)^2 + y^2 - 1) \} \\
&= \inf_{(x,y)} \{y + \lambda_1 (x^2-2x+1) - \lambda_1 (x^2 +2x +1) \} \\
&= \inf_{(x,y)} \{y - 4\lambda_1 x \}
\end{align*}
Dès lors $q(\lambda) = -\infty$.

Supposons maintenant que $\lambda_1 + \lambda_2 \ne 0$. Si $L(x,y,\lambda)$ est strictement convexe par rapport à $(x,y)$, nous pouvons calculer la valeur de $q(\lambda)$ en posant $\nabla (x,y) L(x,y,\lambda) = 0$.

Or, nous avons
$$
\nabla (x,y) L(x,y,\lambda) =
\begin{pmatrix}
2\lambda_1 (x-1) + 2\lambda_2 (x+1) \\
1 + 2\lambda_1 y + 2\lambda_2 y
\end{pmatrix}
$$
et
$$
\nabla^2_{(x,y)} L(x,y,\lambda) =
\begin{pmatrix}
2\lambda_1 + 2\lambda_2 & 0 \\
0 & 2\lambda_1 + 2\lambda_2
\end{pmatrix}
$$
$\det\left(\nabla^2_{(x,y)} L(x,y,\lambda)\right) = 4(\lambda_1+\lambda_2)^2 > 0$.

Si $\lambda_1+\lambda_2 < 0$, $L$ est strictement concave et $q(\lambda) = -\infty$. En effet, en posant $x = 0$,
\begin{align*}
q(\lambda) \,|\,_{x = 0} &= \inf_{y} \{y + \lambda_1 y^2 + \lambda_2 y^2 \} \\
 &= \inf_{y} \{y + (\lambda_1+\lambda_2) y^2 \}
\end{align*}

Si $\lambda_1+\lambda_2 > 0$, $L$ est strictement convexe et $q(\lambda)$ atteint son minimum au zéro du gradient.

En résumé, si $\lambda_1 + \lambda_2 \leq 0$, $q(\lambda) = -\infty$. Autrement $q(\lambda)$ peut être fini.

Supposons $\lambda_1 + \lambda_2 > 0$. $\nabla (x,y) L(x,y,\lambda) = 0$ si et seulement si
$$
\begin{cases}
x = \frac{\lambda_1 - \lambda_2}{\lambda_1 + \lambda_2} \\
y = \frac{-1}{2\lambda_1 + 2\lambda_2}
\end{cases}
$$
En substituant dans $q(\lambda)$, nous obtenons
\begin{align*}
q(\lambda) &=
-\frac{1}{2(\lambda_1 + \lambda_2)} + \lambda_1 \left( \left( \frac{\lambda_1 - \lambda_2}{\lambda_1 + \lambda_2}-1 \right)^2 + \frac{1}{4(\lambda_1 + \lambda_2)^2} - 1 \right) + \lambda_2 \left( \left(\frac{\lambda_1 - \lambda_2}{\lambda_1 + \lambda_2} + 1\right)^2 + \frac{1}{4(\lambda_1 + \lambda_2)^2} - 1 \right) \\
&= -\frac{1}{2(\lambda_1 + \lambda_2)} + (\lambda_1 + \lambda_2) \left( \frac{1}{4(\lambda_1 + \lambda_2)^2} - 1 \right) + 4\frac{\lambda_1\lambda_2^2}{(\lambda_1 + \lambda_2)^2} + 4\frac{\lambda_1^2\lambda_2}{(\lambda_1 + \lambda_2)^2} \\
&= -\frac{1}{2(\lambda_1 + \lambda_2)} + \frac{1}{4(\lambda_1 + \lambda_2)} - (\lambda_1 + \lambda_2)  + 4\lambda_1\lambda_2 \frac{\lambda_1+\lambda_2}{(\lambda_1 + \lambda_2)^2} \\
&= -\frac{1}{4(\lambda_1 + \lambda_2)} + \frac{-(\lambda_1 + \lambda_2)^2+4\lambda_1\lambda_2}{\lambda_1 + \lambda_2} \\
&= -\frac{1}{4(\lambda_1 + \lambda_2)} + \frac{-\lambda_1^2 - \lambda_2^2+2\lambda_1\lambda_2}{\lambda_1 + \lambda_2} \\
&= -\frac{1}{4(\lambda_1 + \lambda_2)}-\frac{(\lambda_1 - \lambda_2)^2}{\lambda_1 + \lambda_2} \\
&= \frac{-1-4(\lambda_1 - \lambda_2)^2}{4(\lambda_1 + \lambda_2)}
\end{align*}

Dès lors $q(\lambda) < 0$, mais $\sup_{\lambda} q(\lambda) = 0$ comme quand $\lambda_1 = \lambda_2 \rightarrow +\infty$, $q(\lambda) \rightarrow 0$. La dualité forte tient, mais asymptotiquement.

Considérons la variante
\begin{align*}
\min_{x,y} \ & y \\
\mbox{s.t. } & (x-1)^2 + y^2 \leq 1 \\
& (x+1)^2 + y^2 \leq 1
\end{align*}

Puisque $y^2 \geq 0$, nous déduisons qu'une condition nécessaire de faisabilité est
$$
\begin{cases}
(x-1)^2 \leq 1 \\
(x+1)^2 \leq 1
\end{cases}
$$

Mais si $x \ne 0$, $(x-1)^2$ ou $(x+1)^2$ est strictement plus grand que 1, aussi une des conditions est violée. Nous devons dès lors avoir $x = 0$ et $y = 0$.

Par conséquent, l'ensemble $\{ (x,y) \,|\, (x-1)^2 + y^2 < 1,\ (x+1)^2 + y^2 < 1 \}$ est vide, et la condition de qualification de contrainte de Slater n'est pas satisfaite.

Il est possible de montrer que la dualité forte tient à nouveau asymptotiquement.

Notons tout d'abord que l'expression du lagrangien demeure la même. On a déjà étudié le cas $\lambda_1 = \lambda_2 = 0$. mais nous devons avoir $\lambda_1 + \lambda_2 > 0$, et dès lors $q(\lambda) < 0$. Mais si nous posons $\lambda_1 = \lambda_2$ et prenons la limite comme $\lambda_1 \rightarrow +\infty$, nous avons, comme précédemment,
$$
\lim_{\lambda_1 \rightarrow +\infty} q(\lambda_1,\lambda_1) = 0.
$$
En d'autres mots, nous ne pouvons trouver une solution duale finie, mais la dualité forte tient encore.

Dans les deux cas, nous ne pouvons pas trouver une solution duale finie en résolvant les conditions KKT.

#### Exemple 2

Considérons le problème
\begin{align*}
\min\ & x+y \\
\mbox{t.q. } & y \geq 0 \\
& y \leq x^3
\end{align*}
La seconde contrainte implique $x \geq 0$, et dès lors, la solution est trivialement $(0,0)$.

La fonction duale lagrangienne est
$$
q(\lambda) = \inf_{x,y} \{ x + y - \lambda_1 y + \lambda_2 (y-x^3) \} = \inf_{x,y} \{ x(1 - x^2 \lambda_2) + y (1+\lambda_2 - \lambda_1) \}
$$

Alors
$$
\nabla_{(x,y)} L(x,y, \lambda) =
\begin{pmatrix}
    1-3x^2\lambda_2 \\
    1-\lambda_1+\lambda_2
\end{pmatrix}
$$

Si $1-\lambda_1+\lambda_2 \ne 0$, il n'y a pas de solution à $\nabla_{(x,y)} L(x,y, \lambda) = 0$, et il est aisé de voir que $q(\lambda) = -\infty$.

Si $1-\lambda_1+\lambda_2 = 0$,
$$
L(x,y, \lambda) = x(1-x^2\lambda_2)
$$
et à nouveau $q(\lambda) = -\infty$.

Dès lors, il y a un saut de dualité infini.

La LICQ ne tient pas comme, en $(0,0)$, les deux contraintes sont actives et la matrice jacobienne des contraintes est
$$
J(x,y) =
\begin{pmatrix}
0 & -1 \\
-3x^2 & 1 \\
\end{pmatrix}
$$
Dès lors,
$$
J(0,0) =
\begin{pmatrix}
0 & -1 \\
0 & 1 \\
\end{pmatrix}
$$

Le système KKT est
\begin{align*}
1 - 3x^2\lambda_2 &= 0 \\
1-\lambda_1+\lambda_2 &= 0 \\
-y &\leq 0 \\
y-x^3 &\leq 0 \\
\lambda_1(-y) &= 0 \\
\lambda_2 (y-x^3) &= 0 \\
\lambda_1, \lambda_2 &\geq 0
\end{align*}

La première égalité implique $\lambda_2 \ne 0$ et dès lors $y = x^3$. La première égalité implique également $x \ne 0$, aussi $y \ne 0$ et $\lambda_1 = 0$. Dès lors, de la seconde égalité, $\lambda_2 = -1$, contredisant $\lambda_2 \geq 0$. Dès lors, le système n'a pas de solution.

### Exemple 3

Considérons le problème convexe
\begin{align*}
\min_{x,y}\ & e^{-x} \\
\mbox{t.q. } & \frac{x^2}{y} \leq 0 \\
& x, y \geq 0.
\end{align*}

La contrainte
$$
\frac{x^2}{y} \leq 0
$$
n'est satisfaite que pour $x = 0$ et $y > 0$.

L'ensemble réalisable est donc $\{ (x,y) \,|\, x = 0,\ y > 0\}$, qui est convexe.

De même, la fonction objectif est convexe, et nous sommes donc en présence d'un problème convexe.

Cependant, comme aucun point tel que $x > 0$ n'est réalisable, la condition de Slater n'est pas satisfaite pour les contraintes telles que formulées.

Comme $x = 0$ pour tout point réalisable, la fonction objectif vaut $1$ quel que soit le point réalisable considéré, et donc toute solution réalisable est également optimale, avec comme valeur optimale la valeur 1.

Le lagrangien est
$$
L(x,y,\lambda) = e^{-x} + \lambda_1 x^2/y - \lambda_2 x - \lambda_3 y
$$
et le problème dual est
\begin{align*}
\max_{\lambda}\ & q(\lambda) = \inf_{x, y} L(x,y,\lambda) \\
\mbox{t.q. } & \lambda \geq 0.
\end{align*}

Afin d'évaluer $q(\lambda)$, cherchons $(x,y)$ tel que
$$
\nabla_{x,y} L(x,y,\lambda) = \begin{pmatrix}
-e^{-x} + 2\lambda_1 x/y - \lambda_2 \\
-\lambda_1 x^2/y^2 - \lambda_3 
\end{pmatrix} = 0
$$

Si $\lambda_3 \ne 0$, $L(0,y,\lambda) = 1-\lambda_3 y$ et $q(\lambda) = -\infty$, en faisant tendre $y$ vers $+\infty$. Observons également que pour $x = 0$, la seconde composante du gradient du lagrangien vaut $-\lambda_3$ et donc ne peut s'annuler.

Considérons à présent le cas $\lambda_3 = 0$. Dans ce cas, la seconde composante du gradient du Lagrangien est nulle pour tout $x = 0$.

Si $\lambda_1 \ne 0$, $\lambda_1 > 0$, et nous devons avoir $x = 0$ et $\lambda_2 = 1$ afin de trouver une racine du gradient du lagrangien. Si $\lambda_2 \ne 1$, alors $q(\lambda)$ n'a pas d'extremum fini comme alors la première composante de $\nabla_{x,y} L(x,y,\lambda)$ devient, pour $x = 0$, $-1 -\lambda_2$. En prenant $y < 0$ et en faisant tendre $x$ vers $+\infty$, nous obtenons $q(\lambda) = -\infty$. La même observation tient pour $\lambda_2 = 1$, et donc la racine du gradient ne correspond pas à un minimum global.

Si $\lambda_1 = 0$, $L(x,y,\lambda) = e^{-x} - \lambda_2 x$. En faisant tendre $x$ vers $+\infty$, nous observons que $q(\lambda) = 0$ si $\lambda_2 = 0$, $-\infty$ sinon.

Par conséquent, le maximum de $q(\lambda)$ vaut 0, atteint avec $\lambda = 0$.

Il y a dès lors un saut de dualité de 1.

Il arrive de conserver les contraintes de non-négativité lors de la construction du lagrangien.
Pour l'exemple, nous avons alors
$$
L(x,y,\lambda) = e^{-x} + \lambda x^2/y
$$
et
$$
q(\lambda) = \inf_{x,y \geq 0} L(x,y,\lambda)
$$

Dans ce cas, si $\lambda \geq 0$, $L(x,y,\lambda) > 0$ et $q(\lambda) = 0$, par exemple en prenant $x = t$ et $y = t^3$, comme alors
$$
L(x,y,\lambda) = e^{-t} + \lambda t^2/t^3 = e^{-t} + \lambda 1/t
$$
et il suffit de faire tendre $t$ vers $+\infty$.

Le saut de dualité est dès lors à nouveau de 1.
